```python
# The task is to split data into training data and test data 80%/20%
```

```python
import pandas as pd
import shutil
from pathlib import Path
```

```python
df = pd.read_csv("../data/metadata_with_group.csv")     # works if the jupiter notebook in a different folder from data
```

```python
# malignant diseases: BCC, MEL, SCC
malignants = df[df["diagnostic"].isin(["BCC", "MEL", "SCC"])]
df_BCC = df[df["diagnostic"] == "BCC"]
df_SCC = df[df["diagnostic"] == "SCC"]
df_MEL = df[df["diagnostic"] == "MEL"]
df_malignant = pd.concat([df_MEL[:int(len(df_MEL)*0.8)], df_SCC[:int(len(df_SCC)*0.8)], df_BCC[:int(len(df_BCC)*0.8)]]).sort_values(by="patient_id")
len(df_malignant)/len(malignants)       # check that 80% is the training data
```

```python
# benign Diseases: ACK, NEV, SEK
benigns = df[df["diagnostic"].isin(["ACK", "NEV", "SEK"])]
df_ACK = df[df["diagnostic"] == "ACK"]
df_SEK = df[df["diagnostic"] == "SEK"]
df_NEV = df[df["diagnostic"] == "NEV"]
df_benign = pd.concat([df_NEV[:int(len(df_NEV)*0.8)], df_SEK[:int(len(df_SEK)*0.8)], df_ACK[:int(len(df_ACK)*0.8)]]).sort_values(by="patient_id")
len(df_benign)/len(benigns) # check that 80% is the training data
```

```python
training_df = pd.concat([df_malignant, df_benign]).sort_values(by="patient_id")
len(training_df)/len(df)    # check that 80% is the training data
```

```python
# FOR TRAINING SET ONLY
training_path = Path("training data")
raw_path = Path("../data/imgs/")
training_path.mkdir(parents=True, exist_ok=True)
for name in training_df["img_id"]:
        shutil.copyfile(f"{raw_path}/{name}", f"{training_path}/{name}")
```

```python
training_df.to_csv(f"training_data.csv")        # convert to csv
```

```python
test_df = df[~df["img_id"].isin(training_df["img_id"])].sort_values(by="patient_id")
```

```python
# FOR TEST SET ONLY
test_path = Path("test data")
raw_path = Path("../data/imgs/")
test_path.mkdir(parents=True, exist_ok=True)
for name in test_df["img_id"]:
        shutil.copyfile(f"{raw_path}/{name}", f"{test_path}/{name}")
```

```python
test_df.to_csv(f"test_data.csv")
```
